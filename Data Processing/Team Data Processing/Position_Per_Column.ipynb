```python
import pandas as pd
import numpy as np

# File path
file_path = r'C:\Users\gurpr\OneDrive\Documents\New Projects\Euros Match Predictor\Data\Final Data\Mean_Country_Data.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Identify numerical and float columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Initialize an empty list to store the output for each column
outputs = []

# Loop through each numerical column
for col in numeric_cols:
    # Group by 'Position' and calculate the mean for the current numerical column
    grouped_means = df.groupby('Position')[col].mean()
    
    # Sort by the mean and get the top positions
    sorted_means = grouped_means.sort_values(ascending=False)
    
    # Get the top position and its value
    top_position = sorted_means.index[0]
    top_value = sorted_means.iloc[0]
    
    top_positions_list = [top_position]
    
    # Check for the second position and its value
    if len(sorted_means) > 1:
        second_position = sorted_means.index[1]
        second_value = sorted_means.iloc[1]
        
        # Include the second position if its value is at least 80% of the top position's value
        if second_value >= 0.8 * top_value:
            top_positions_list.append(second_position)
    
    # Format the output for the current column
    output = f"'{col}':{top_positions_list},"
    
    # Add the formatted output to the list
    outputs.append(output)

# Print all outputs
for output in outputs:
    print(output)

```
